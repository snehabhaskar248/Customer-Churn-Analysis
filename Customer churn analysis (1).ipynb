Replacing blanks with 0 as tenure is 0 and no total charges are recorded

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

defining a function to convert values 0 and 1 to yes and No in SeniorCitizen column

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "No"
    
df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
df.tail()
```

```python
ax = sns.countplot(x = 'Churn', data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn", fontsize=15)
plt.show()
```

```python
gb = df.groupby("Churn").agg({'Churn':'count'})
plt.pie(gb['Churn'], labels=gb.index, autopct="%1.2f%%")
plt.title("Percentage of Churned Customers", fontsize=15)
plt.show()
```

From above Pie Chart we can conclude that 26.54% of our customers have churned out.
Now let's explore the reason behind it.

```python
A1 = sns.countplot(x = 'gender', data = df, hue='Churn')
A1.bar_label(A1.containers[0])
plt.title("Churn by gender", fontsize=15)
plt.show()
```

```python
A2 = sns.countplot( x = 'SeniorCitizen' , data = df , hue = 'Churn')
plt.title("Churn by SeniorCitizen", fontsize=15)
plt.show()

```

```python
data = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)
percent = data.divide(data.sum(axis=1), axis=0)

# Plot
ax = percent.plot(kind='bar', stacked=True, colormap='viridis')
for i, senior in enumerate(percent.index):
    for j, churn in enumerate(percent.columns):
        p = percent.loc[senior, churn]
        if p > 0.03:
            y = percent.iloc[i, :j].sum() + p / 2
            ax.text(i, y, f'{p*100:.1f}%', ha='center', va='center', color='white', weight='bold')
plt.title("Churn by SeniorCitizen (% to Total)", fontsize=15)
plt.xlabel("SeniorCitizen")
plt.ylabel("Percent to Total")
plt.legend(title='Churn')
plt.tight_layout()
plt.show()
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

fig, axes = plt.subplots(3, 3, figsize=(18, 14))
axes = axes.flatten()

for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, ax=axes[i], hue = df["Churn"])
    axes[i].set_title(col)
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

The data shows that churn is significantly lower among customers who have phone service, internet service, and access to various protections or streaming options across all categories. For most service types (e.g., OnlineSecurity, DeviceProtection, StreamingTV), non-churning customers substantially outnumber churners. "No internet service" and "No phone service" categories have a smaller overall count and relatively lower churn rates compared to those with active services.
